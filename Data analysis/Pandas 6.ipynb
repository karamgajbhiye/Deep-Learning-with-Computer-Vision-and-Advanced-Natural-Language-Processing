```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
url1 = 'https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv'
url2 = 'https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv'
url3 = 'https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv'
url4 = 'https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv'
```

```python
users = pd.read_csv(url1)
sessions = pd.read_csv(url2)
products = pd.read_csv(url3)
transactions = pd.read_csv(url4)
```

```python
#Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)
```

```python
transactions.merge(users,on='UserID',how='left') # ortransactions.merge(users,how='left')
```

```python
pd.merge(transactions,users,how='left')
```

```python
#Which transactions have a UserID not in users?
```

```python
transactions[~transactions['UserID'].isin(users['UserID'])]
```

```python
#Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)
```

```python
users.merge(transactions,how='inner',on='UserID')
```

```python
#Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
```

```python
users.merge(transactions,how='outer')  #pd.merge(users,transactions,how='outer')
```

```python
#Determine which sessions occurred on the same day each user registered
```

```python
users.merge(sessions,how='inner',left_on=['UserID','Registered'],right_on=['UserID','SessionDate'])
```
