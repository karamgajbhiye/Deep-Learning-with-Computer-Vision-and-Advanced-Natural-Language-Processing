```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
X_Data=iris.data
```

```python
#X_Data
```

```python
y_target = iris.target
```

```python
#y_target
```

```python
#PCA(n_components=3).fit_transform(iris.iloc[:,0:4].values)
```

```python
pca = PCA(n_components=3)
X_components = pca.fit_transform(X_Data)
```

```python
#X_components
```

```python
fig = plt.figure(1,figsize=(9,9))
axis = Axes3D(fig,elev=-148, azim=155)
axis.scatter(X_components[:, 0], X_components[:, 1], X_components[:, 2],c=y_target,cmap=plt.cm.rainbow_r)
axis.set_title('IRIS DATA')
axis.set_xlabel('X')
axis.set_ylabel('Y')
axis.set_zlabel('Z')
axis.legend()
plt.show()
```
